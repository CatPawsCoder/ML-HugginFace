```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
ARTICLE = """ UrFU offers a variety of Master's degree programs taught in English by leading professors and researchers of national and international level.

For the academic year 2023-24, there are 13 Master's degree programs taught fully in English and 3 Master's degree programs taught partially in English (at least 50% of academic hours).

Eligibility criteria: Bachelor's degree or other equivalent qualification.

Application period: February 15 to October 28
Enrollment deadline (for fee-paying students): October 28
Submit your documents via Online Application Profile
"""
print(summarizer(ARTICLE, max_length=118, min_length=30, do_sample=False))
```
